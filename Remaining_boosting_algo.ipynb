```python
from sklearn.cluster import KMeans
from sklearn.preprocessing import StandardScaler
from kmodes.kmodes import KModes
import matplotlib.pyplot as plt
```

```python
df.fillna(value=0, inplace=True)
```

```python
clustering=pd.get_dummies(df,drop_first=True)
```

```python
# Elbow curve to find optimalk

cost=[]
K=range(1,6)
for k in list(K):
    kmode=KModes(n_clusters=k,init='random',n_init=5, verbose=1)
    kmode.fit_predict(clustering)
    cost.append(kmode.cost_)
```

```python
plt.plot(K,cost,'x-')
plt.xlabel('No. of clusters')
plt.ylabel('Cost')
plt.title('Elbow Curve')
plt.show()
```

```python
kmode=KModes(n_clusters=3,init='random',n_init=5,verbose=1)
clus=kmode.fit_predict(clustering)
```

```python
import numpy as np

from sklearn.datasets import load_iris
from sklearn.preprocessing import StandardScaler
from scipy.cluster.hierarchy import dendrogram, linkage
from sklearn.cluster import AgglomerativeClustering
```

```python
iris=load_iris()
X=iris.data
```

```python
# feature scaling
scaler =StandardScaler()
X_scaled=scaler.fit_transform(X)
```

```python
# perform hierarchical clustering

Z=linkage(X_scaled,method='ward') # u can chhose different methods : 'single', 'complete' , 'average', 'ward'
```

```python
# plot the dendrogram

plt.figure(figsize=(10,7))
dendrogram(Z,truncate_mode='level',p=3) # truncate_mode='level' for easier visualization 
plt.title('Hierarchical Clustering Dendrogram')
plt.xlabel('Sample index')
plt.ylabel('Distance')
plt.show()
```

```python
# fir the hiererchial clustering model

hier_clust=AgglomerativeClustering(n_clusters=3,metric='euclidean', linkage='ward')
hier_clust.fit(X_scaled)
```

```python
# add cluster labels to the dataset

labels=hier_clust.labels_
print(f'Cluster labels:\n{labels}')
```
