```python
Linear Regression Relies on several key assumptions for its validity. Here are the main assumption:
Linearity: The relationship between the dependent variable (Y) and the independent variable (X) is linear. This means tha the
change in Y is proportional to change in X.

Normality: The residuals are normally distributed. This means that the distribution of the residuals should resembel a bell curve when plotted.

No Perfect Multicollinearity: There should be no exact linear relationship between the independent variables. In other words, one independent variable should not be a perfect linear combination of the others.

Homoscedasticity: Also known as constant variance, this assumption states that the variance of the residuals (the difference between the observed and predicted values)
is constant accross all levels of the independent variable (s ). In simpler terms, the spread of the residuals should remain the same throughout the range of X.

No Autocorrelation: The residuals are not correlated with each other. This means that there is no pattern in the residuals when plotted
against time or the order in which they were collected.
```

```python
MODEL BUILDING
```

```python
import statsmodels.api as sm
```

```python
# statsmodel statistical model -- OLS (Ordinary Least Square)
# sklearn - linear regression
```

```python
model=sm.OLS(ytrain,xtrain).fit()  # r squared value should be minimum
model.summary()
```

```python
PREDICTION
```

```python
pred_train=model.predict(xtrain)
pred_test=model.predict(xtest)
```

```python
MODEL METRICES
```

```python
from sklearn.metrics import mean_squared_error
import numpy as np
```

```python
train_rmse=np.sqrt(mean_squared_error(ytrain,pred_train))
test_rmse=np.sqrt(mean_squared_error(ytest,pred_test))
print(f"Train_Rmse",train_rmse)
print(f"Test_Rmse",test_rmse)
```
