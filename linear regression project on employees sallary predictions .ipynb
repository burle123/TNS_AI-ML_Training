```python
from sklearn.model_selection import train_test_split
```

```python
x= df.drop("Salary", axis=1) #independent variable
y= df["Salary"]
```

```python
import statsmodels.api as sm
```

```python
x_with_constant=sm.add_constant(x)
x_with_constant
```

```python
y
```

```python
xtrain,xtest,ytrain,ytest= train_test_split(x_with_constant,y,test_size=0.3,random_state=30)
```

```python
xtest.head()
```

```python
xtrain.index
```

```python
model=sm.OLS(ytrain,xtrain).fit()
model.summary()
```

```python
#PREDICTION
```

```python
pred_train=model.predict(xtrain)
pred_test=model.predict(xtest)
```

```python
#MODEL METRICES
```

```python
from sklearn.metrics import mean_squared_error
import numpy as np
```

```python
train_rmse=np.sqrt(mean_squared_error(ytrain,pred_train))
test_rmse=np.sqrt(mean_squared_error(ytest,pred_test))
print(f"Train_Rmse",train_rmse)
print(f"Test_Rmse",test_rmse)
```
