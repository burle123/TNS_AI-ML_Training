```python
#Step 2: Import libraries
from transformers import pipeline
import pandas as pd

# Step 3: Load pre-trained language model (GPT-2)
generator = pipeline("text-generation", model="gpt2")

print("Model Loaded Successfully")

#Step 4: Define prompts and parameters

prompts = [

"Artificial Intelligence is",
"Machine learning helps in",
"Data science is important because"
]

temperatures=[0.5,0.8,1.2]
max_length=10

# Step 5: generate text and store results
results=[]

for prompt in prompts:
  for temp in temperatures:
    output = generator(prompt,
                       max_length=max_length,
                       temperature=temp,
                       num_return_sequences=1)

    generated_text=output[0]['generated_text']
    results.append({
        "Prompt":prompt,
        "Temperature":temp,
        "Max_Length":max_length,
        "Generated_Output":generated_text,
        "Output_Length":len(generated_text)
        })

#step 6: convert results to table(dataframe)
df= pd.DataFrame(results)
print(df)

#step 7: save result to excel file
df.to_excel("LLM_Text_Generation_Result.xlsx", index=False)
print("\nFile saved as LLM_Text_Generation_Result.xlsx")

#step 8: display dataframe nicely in colab
df
```
