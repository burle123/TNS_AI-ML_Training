```python
model=tf.keras.models.Sequential(
    [
        tf.keras.layers.Dense(8,activation='sigmoid',input_dim=(784)),
        tf.keras.layers.Dense(10,activation='softmax')
    ]
)
```

```python
model.compile(optimizer='adam',loss='categorical_crossentropy',metrics=['accuracy'])
```

```python
model.summary()
```

```python
df
```

```python
y_train_encod=pd.get_dummies(y_train)
```

```python
history=model.fit(df.values,y_train_encod,batch_size=5,epochs=5,validation_data=(x_test.reshape(10000,-1),pd.get_dummies(y_test)))
```

```python
plt.plot(history.history['accuracy'])
plt.plot(history.history['val_accuracy'])
```
