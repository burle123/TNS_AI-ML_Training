```python
# FINAL OPTIMIZED ENTERTAINMENT CHATBOT

import random
import nltk
import string
import numpy as np

from nltk.stem import WordNetLemmatizer
from nltk.corpus import stopwords
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.linear_model import LogisticRegression

# Download required resources (run once)
nltk.download('punkt')
nltk.download('stopwords')
nltk.download('wordnet')

# Initialize tools
lemmatizer = WordNetLemmatizer()
stop_words = set(stopwords.words("english"))

# KNOWLEDGE BASE (EXPANDED)
intents = {
    "intents": [
        {
            "tag": "greeting",
            "patterns": [
                "hi", "hello", "hey", "what's up",
                "good morning", "good evening", "hi there",
                "hey buddy", "hello chatbot", "yo"
            ],
            "responses": [
                "Hey! Ready for some fun? 😄",
                "Hello! Want jokes, movies, or games?",
                "Hi there! Let's have some fun 🎉"
            ]
        },
        {
            "tag": "jokes",
            "patterns": [
                "joke", "jokes", "tell me a joke", "make me laugh",
                "funny", "say something funny", "crack a joke",
                "i want to laugh", "give me humor"
            ],
            "responses": [
                "Why don’t programmers like nature? Too many bugs! 😂",
                "Why did the computer go to therapy? It had too many issues 😆",
                "I told my laptop I needed a break… it froze 😜"
            ]
        },
        {
            "tag": "movies",
            "patterns": [
                "movie", "movies", "movie suggestion", "recommend a movie",
                "suggest films", "what movie should i watch",
                "any good movies", "cinema suggestions"
            ],
            "responses": [
                "If you like thrillers, try Inception 🎬",
                "Comedy fan? The Hangover never disappoints 😄",
                "For inspiration, watch The Pursuit of Happyness 🌟"
            ]
        },
        {
            "tag": "music",
            "patterns": [
                "music", "songs", "music suggestion", "recommend songs",
                "what should i listen", "i love music",
                "play music", "song ideas"
            ],
            "responses": [
                "Try listening to lo-fi beats for relaxation 🎧",
                "Feeling energetic? EDM is the way to go 🔥",
                "Old melodies or new hits – music heals everything 🎶"
            ]
        },
        {
            "tag": "mood_happy",
            "patterns": [
                "i am happy", "feeling good", "excited",
                "i feel great", "i am आनंद", "i am feeling awesome"
            ],
            "responses": [
                "That's awesome! Keep smiling 😄",
                "Happiness looks great on you 🌈"
            ]
        },
        {
            "tag": "mood_sad",
            "patterns": [
                "i am sad", "feeling low", "depressed",
                "not feeling good", "i feel bad",
                "i am upset", "feeling down"
            ],
            "responses": [
                "I'm here for you 💙 Things will get better.",
                "Take a deep breath. Want a joke or a game?"
            ]
        },
        {
            "tag": "game",
            "patterns": [
                "play a game", "game time", "let's play",
                "start game", "i want to play", "game"
            ],
            "responses": [
                "Let's play a number guessing game!"
            ]
        },
        {
            "tag": "goodbye",
            "patterns": [
                "bye", "exit", "quit", "see you",
                "goodbye", "bye bye"
            ],
            "responses": [
                "Goodbye! Stay entertained 😎",
                "See you again! 🎉"
            ]
        }
    ]
}

# PREPROCESSING (FIXED)
def preprocess(text):
    text = text.lower().strip()

    # remove quotes
    text = text.replace("'", "").replace("*", "")

    tokens = nltk.word_tokenize(text)

    tokens = [
        lemmatizer.lemmatize(word)
        for word in tokens
        if word not in string.punctuation and word not in stop_words
    ]

    return " ".join(tokens)


# TRAINING DATA
corpus = []
labels = []

for intent in intents["intents"]:
    for pattern in intent["patterns"]:
        corpus.append(preprocess(pattern))
        labels.append(intent["tag"])


# VECTORIZATION
vectorizer = TfidfVectorizer(ngram_range=(1, 2))
X = vectorizer.fit_transform(corpus)


# MODEL
model = LogisticRegression(max_iter=200)
model.fit(X, labels)


# CONTEXT MEMORY
context = {
    "last_intent": None,
    "game_number": None
}


# CHATBOT FUNCTION
def chatbot(user_input):
    processed = preprocess(user_input)
    vector = vectorizer.transform([processed])

    prob = model.predict_proba(vector)[0]
    confidence = np.max(prob)
    intent_tag = model.predict(vector)[0]

    print(f"[DEBUG] Intent: {intent_tag}, Confidence: {confidence:.2f}")

    # GAME CONTINUE
    if context["last_intent"] == "game" and user_input.isdigit():
        guess = int(user_input)
        if guess == context["game_number"]:
            context["last_intent"] = None
            return "Correct! You win 🎉"
        else:
            return "Wrong! Try again."

    # LOW CONFIDENCE RESPONSE
    if confidence < 0.25:
        return random.choice([
            "Not sure I got that, but I can tell jokes, suggest movies or music!",
            "Try asking for jokes, movies, or games",
            "I'm still learning! Ask me something fun"
        ])

    # GAME START
    if intent_tag == "game":
        context["game_number"] = random.randint(1, 5)
        context["last_intent"] = "game"
        return "I picked a number between 1 and 5. Guess it!"

    context["last_intent"] = intent_tag

    for intent in intents["intents"]:
        if intent["tag"] == intent_tag:
            return random.choice(intent["responses"])


# CHAT LOOP
print("Entertainment AI Chatbot is live! (type 'exit' to stop)\n")

while True:
    user = input("You: ")
    if user.lower() in ["exit", "quit"]:
        print("Bot: Goodbye! Stay Happy!")
        break

    print(f"Bot: {chatbot(user)}")
```

```python
def data_analyst_bot():
  data = input("Paste dataset (table or text):\n")

  prompt = f"""
  Analyze the data and provide:
  - Key insights
  - Trends
  - Anomalies
  - Business recommendations
  """
  response_text = chatbot(prompt+data)
  print("\nInsights:\n", response_text)

data_analyst_bot()
```
