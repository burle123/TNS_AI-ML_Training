```python
# ------------------------------------------------------------
# STEP 1: CREATE A SMALL INSTRUCTION DATASET
# ------------------------------------------------------------

data = [
    {
        "instruction": "What is Artificial Intelligence?",
        "input": "",
        "output": "Artificial Intelligence is the simulation of human intelligence in machines."
    },
    {
        "instruction": "Define Machine Learning",
        "input": "",
        "output": "Machine Learning is a subset of AI that enables systems to learn from data."
    },
    {
        "instruction": "What is NLP?",
        "input": "",
        "output": "Natural Language Processing focuses on interaction between computers and human language."
    }
]

print("Number of samples:", len(data))
print("First sample:", data[0])
```

```python
# ------------------------------------------------------------
# STEP 2: EXPLORE AND CLEAN DATA
# ------------------------------------------------------------

for sample in data:
    sample["Instruction:"]= sample["instruction"].strip()
    sample["output"]= sample["output"].strip()

lengths = [len(sample["output"].split()) for sample in data]
print(lengths)
print("Average length:", sum(lengths)/len(lengths))
```

```python
# ------------------------------------------------------------
# STEP 3: TOKENIZATION PREVIEW
# ------------------------------------------------------------

from transformers import AutoTokenizer

tokenizer = AutoTokenizer.from_pretrained("gpt2")
tokens= tokenizer(data[0]["instruction"])
#attension mask
#1 -> real token (important, attend to it)
#0 -> padding token (ignore it)

print(tokenizer.tokenize(data[0]["instruction"]))
print(tokens)
print("Token length:", len(tokens["input_ids"]))
```

```python
# ------------------------------------------------------------
# STEP 4: FORMAT FOR TRAINING
# ------------------------------------------------------------
def format_sample(sample):
  return f"Instruction: {sample['instruction']}\nResponse:{sample['output']}"

formatted_data= [format_sample(s) for s in data]
print("formatted sample:\n", formatted_data[0])
```

```python
# ------------------------------------------------------------
# STEP 5: COMPARE MULTIPLE LLM OUTPUTS
# ------------------------------------------------------------

from transformers import pipeline

generator1 = pipeline("text-generation", model="gpt2")
generator2= pipeline("text-generation", model="distilgpt2")

prompt= "Explain Artificial Intelligence in simple terms."

print("\nModel 1 output")
print(generator1(prompt, max_length=50)[0]["generated_text"])

print("\nModel 2 output")
print(generator2(prompt, max_length=50)[0]["generated_text"])
```
