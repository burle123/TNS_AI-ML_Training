```python
# Using OpenAI / Hugging Face APIs & Building Gen-AI Applications

# 1. Introduction to Generative AI APIs
# Generative AI APIs are tools that allow developers to access powerful pre-trained models through simple requests. Instead of building complex models from scratch, developers can send input (prompt) to the API and receive intelligent output such as text, summaries, or conversations.
# Organizations like OpenAI and Hugging Face provide these services. These platforms host large language models trained on vast datasets, enabling applications to perform human-like text generation.
# The main advantage is speed and efficiency, as developers can focus on building applications rather than training models.

# 2. Working Principle of AI APIs
# Generative AI APIs follow a simple workflow:
# •	The user provides input (prompt)
# •	The API processes the input using a trained model
# •	The system generates a response based on patterns learned from data
# This process relies on Natural Language Processing (NLP), where the model understands context, intent, and language structure to produce meaningful output.
# These APIs are typically accessed through HTTP requests and return structured responses, often in JSON format.

# 3. OpenAI API – Conceptual Understanding
# The OpenAI API provides access to advanced language models capable of:
# •	Text generation
# •	Conversation (chatbots)
# •	Content creation
# •	Question answering
# It is designed for ease of use, allowing developers to integrate AI into applications quickly. The API works by sending a prompt and receiving a response generated by the model.
# It is widely used in:
# •	Virtual assistants
# •	Content writing tools
# •	Coding assistants
# •	Educational platforms

# 4. Hugging Face API – Conceptual Understanding
# Hugging Face is a platform that offers a large collection of pre-trained models for various NLP tasks. It supports both API-based usage and local model deployment.
# Key capabilities include:
# •	Text generation
# •	Summarization
# •	Translation
# •	Sentiment analysis
# One of its strengths is flexibility. Developers can choose from thousands of models depending on their use case and even fine-tune them for specific applications.

# 5. Building Generative AI Applications
# Generative AI applications use these APIs to create intelligent systems that can generate content or interact with users.
# A. Chatbot Applications
# A chatbot is a system that simulates human conversation. It takes user input and responds intelligently using an AI model.
# Core components:
# •	User interface (text or voice)
# •	API integration
# •	Response generation
# Applications include:
# •	Customer support
# •	Educational tutors
# •	Personal assistants
# B. Text Summarization Tools
# Summarization tools convert large amounts of text into shorter, meaningful summaries while preserving key information.
# Types of summarization:
# •	Extractive: Selects important sentences
# •	Abstractive: Generates new sentences
# Use cases:
# •	News summarization
# •	Report analysis
# •	Meeting notes generation
# C. Resume Builder Tools
# Resume generators use AI to create professional resumes based on user input such as skills, experience, and education.
# Features:
# •	Structured formatting
# •	Professional language
# •	Customization for job roles
# These tools help users save time and improve the quality of their resumes.
# 6. Key Differences: OpenAI vs Hugging Face
# •	OpenAI focuses on simplicity and ease of integration through APIs.
# •	Hugging Face provides more flexibility with access to multiple models and customization options.
# •	OpenAI is primarily cloud-based, while Hugging Face supports both cloud and local deployment.
# •	Hugging Face is widely used in research and experimentation, whereas OpenAI is popular for production-ready applications.
# 7. Advantages of Using Generative AI APIs
# •	Eliminates the need for model training
# •	Reduces development time
# •	Provides high-quality outputs
# •	Easily scalable for real-world applications
# •	Enables rapid prototyping of AI solutions
# 8. Challenges and Limitations
# •	Dependency on external APIs
# •	Cost associated with usage
# •	Data privacy concerns
# •	Possibility of incorrect or biased outputs
# •	Limited control over model behaviour (especially in API-only systems)
# 9. Real-World Applications
# Generative AI APIs are widely used in:
# •	Healthcare (report summarization)
# •	Education (AI tutors)
# •	Business (automated customer support)
# •	Content creation (blogs, marketing)
# •	Recruitment (resume screening and generation)
# 10. Future Scope of Generative AI
# Generative AI is evolving rapidly and is expected to:
# •	Improve personalization in applications
# •	Enhance human-AI interaction
# •	Automate complex tasks
# •	Integrate with multimodal systems (text, image, audio)
# It is becoming a core technology in modern software development.
```

```python
#STEP 4 : Chat Function

def chatbot_response(user_input):
    response = model.generate_content(user_input)
    return response.text
```

```python
#STEP 5 : Chat Loop

print("Gemini Entertainment Chatbot is Live!")
print("Type 'exit' to stop\n")

while True:
  user_input = input("You: ")
  if user_input.lower() in ["exit", "quit"]:
      print("Gemini Entertainment Chatbot: Goodbye!")
      break

  answer = chatbot_response(user_input)
  print("Bot: ",answer)
```

```python
#Resume Analyzer

def advanced_resume_analyzer():
  print("\n---Resume Analyzer---")

  resume = input("Paste Resume:\n")
  job_description = input("Paste Job Description:\n")

  prompt = f"""
  You are an ATS System.
  Analyze the resume against the Job Description.

  Output Format:
  1.ATS SCore(0-100)
  2.Matching Skills
  3.Missing Keywords
  4.Strengths
  5.Weaknesses
  6.Improved Resume Points
  7.Final Suggestions

  Resume:
  {resume}

  Job Description:
  {job_description}
  """

  try:
    response = model.generate_content(prompt)
    print("\n---Analysis Result---\n")
    print(response.text)

  except Exception as e:
    print("Error:", e)

advanced_resume_analyzer()
```
