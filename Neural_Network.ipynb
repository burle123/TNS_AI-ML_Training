```python
# A Neural Network (NN) is a system of connected nodes (called neurons) that work together to learn patterns from data and make predictions.

# Just like the human brain:

# Brain → Neurons
# Neural Network → Artificial Neurons



# Basic Structure
# A neural network has 3 main layers:

# Input Layer
# Takes input data (e.g., image pixels, numbers)
# Hidden Layer(s)
# Performs calculations and feature extraction
# Can have multiple layers → called Deep Learning
# Output Layer
# Produces final result (e.g., class label, prediction)



# How It Works (Step-by-Step)
# Input data is fed into the network
# Each neuron:
# Multiplies input with weights
# Adds bias
# Applies activation function
# Output is generated
# Error is calculated
# Network adjusts weights using backpropagation



# Key Components
# 1. Weights
# Importance given to each input
# 2. Bias
# Helps shift output
# 3. Activation Function

# Examples:

# ReLU
# Sigmoid
# Tanh

#Real Life Example:

# Chatbots
# Self-Driving cartesian
# Recommendation System (Netflix,Amazon)

# Advantages:

# learns complex patterns
# works well with large data
# High accuracy_score

# Disadvantages:

# Needs Large Data
# Time-Consuming
# Hard to interpret (Black box)
```

```python
#Step 1 - import libraries

import numpy as np
import pandas as pd

from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler, LabelEncoder
from sklearn.neural_network import MLPClassifier
from sklearn. import accuracy_score, classification_report
metrics
```

```python
#Step 2 - Load Dataset
df = pd.read_csv("/content/Churn_Modelling.csv")
```

```python
#Step 3 - Preprocessing
le = LabelEncoder()
df['Gender'] = le.fit_transform(df['Gender'])
df['Geography'] = le.fit_transform(df['Geography'])
```

```python
#Drop unnecessary column
df = df.drop(['CustomerId','Surname'],axis=1)
```

```python
#Features and Target
X = df.drop('Exited',axis=1)
y = df['Exited']

```

```python
#Step 4 - Train-Test Split
X_train, X_test, y_train, y_test = train_test_split(X,y,test_size=0.2)
#
```

```python
#Step-5 Scaling
scaler = StandardScaler()
X_train = scaler.fit_transform(X_train)
X_test = scaler.transform(X_test)
```

```python
#Step 6 - ANN Model

model = MLPClassifier(hidden_layer_sizes=(20,10), activation = 'relu' ,max_iter=1000)
model.fit(X_train,y_train)
```

```python
#Step 7 - Train
model.fit(X_train,y_train)
```

```python
#Step 8 - Predict
y_pred = model.predict(X_test)
```

```python
#Step 9 - Evaluate
print("Accuracy:", accuracy_score(y_test, y_pred))
print("\nReport:\n", classification_report(y_test, y_pred))
```
