```python
"""
NLP BASICS TEXT CLEANING ONLY (NO TOKENIZATION)

This code demonstrates essential NLP text cleaning steps used before any linguistic processing.

Steps Covered:

1. Lowercasing

2. Removing Numbers

3. Removing Punctuation

4. Removing Emojis & Special Characters

5. Expanding Contractions

6. Removing Extra Spaces

7. Normalizing Repeated Characters
"""

#STEP 8: Import Library

import re

#step 1: simple text
text1 = "WOW!!! NLP in 2025 is sooo COOOOL 😍😍"
text2 = "I can't believe this!!! It's 100% true..."
text3 = "This     sentence     has      extra spaces!!!"
text4 = "Nooooo wayyyyy this works????"

texts= [text1,text2,text3,text4]

#step 2: contraction Expamsion

def expands_contractions(text):
  contractions={
      "can't":"cannot",
      "it's": "it is",
      "i'm": "i am",
      "don't": "do not",
      "won't":"will not"
  }
  for c, e in contractions.items():
    text= re.sub(r"\b"+c+r"\b",e,text)
    return text

#step 3: text cleaning function

def clean_text(text):
  text=text.lower()                                 # Lowercase
  text=expands_contractions(text)                   # expand contractions
  text=re.sub(r'\d+','',text)                       # remove numbers
  text=re.sub(r'[^a-z\s]','',text)                  # Remove punctuations and emojis
  text=re.sub(r'(.)\1{2, }',r'\1\1',text)           # normalize repeated characters
  text= re.sub(r'\s+',' ',text).strip()             # remove extra spaces
  return text
```

```python
#STEP 4: Apply Cleaning

print("TEXT CLEANING OUTPUT:\n")


for text in texts:
    cleaned =clean_text(text)

    print("Original", text)

    print("Cleaned", cleaned)

    print("-"*50)
```
