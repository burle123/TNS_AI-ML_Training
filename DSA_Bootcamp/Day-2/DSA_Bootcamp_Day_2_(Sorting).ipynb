# 🔹 Sorting Algorithms

Sorting is the process of arranging elements in a specific order (ascending or descending).

---

## ✅ Why Sorting?
- Makes searching faster (Binary Search requires sorted data)
- Helps in data organization and analysis
- Widely used in real-world applications (databases, ML, etc.)

---

## 🔸 Types of Sorting Algorithms

### 1. Bubble Sort
- Compares adjacent elements and swaps them if needed
- Repeats until the array is sorted
- Time Complexity: O(n²)

---

### 2. Selection Sort
- Finds the minimum element and places it at correct position
- Reduces unnecessary swaps
- Time Complexity: O(n²)

---

### 3. Insertion Sort
- Builds sorted array step-by-step
- Efficient for small or nearly sorted data
- Time Complexity: O(n²), Best Case: O(n)

---

### 4. Merge Sort
- Uses Divide & Conquer
- Splits array and merges sorted halves
- Time Complexity: O(n log n)

---

### 5. Quick Sort
- Uses pivot to partition array
- Efficient in practice
- Time Complexity: O(n log n), Worst: O(n²)

---

## 🔹 Bubble Sort

- Repeatedly compares adjacent elements
- Swaps if they are in wrong order
- Largest element "bubbles" to the end

### ⏱ Time Complexity:
- Worst: O(n²)
- Best: O(n)

### ✅ Stable: Yes  
### 📦 In-place: Yes

```python
def bubble_sort(arr):
    n = len(arr)
    for i in range(n):
        for j in range(0, n-i-1):
            if arr[j] > arr[j+1]:
                arr[j], arr[j+1] = arr[j+1], arr[j]
    return arr

arr = [64, 34, 25, 12, 22, 11, 90]
print("Bubble Sort:", bubble_sort(arr.copy()))
```


## 🔹 Selection Sort

- Finds the minimum element and places it at correct position
- Reduces number of swaps

### ⏱ Time Complexity:
- O(n²)

### ❌ Stable: No  
### 📦 In-place: Yes

```python
def selection_sort(arr):
    n = len(arr)
    for i in range(n):
        min_idx = i
        for j in range(i+1, n):
            if arr[j] < arr[min_idx]:
                min_idx = j
        arr[i], arr[min_idx] = arr[min_idx], arr[i]
    return arr

arr = [64, 34, 25, 12, 22, 11, 90]
print("Selection Sort:", selection_sort(arr.copy()))
```

## 🔹 Insertion Sort

- Builds sorted array step-by-step
- Inserts element at correct position

### ⏱ Time Complexity:
- Worst: O(n²)
- Best: O(n)

### ✅ Stable: Yes  
### 📦 In-place: Yes

```python
def insertion_sort(arr):
    for i in range(1, len(arr)):
        key = arr[i]
        j = i - 1
        while j >= 0 and arr[j] > key:
            arr[j+1] = arr[j]
            j -= 1
        arr[j+1] = key
    return arr

arr = [64, 34, 25, 12, 22, 11, 90]
print("Insertion Sort:", insertion_sort(arr.copy()))
```

## 🔹 Merge Sort

- Divide & Conquer algorithm
- Splits array and merges sorted parts

### ⏱ Time Complexity:
- O(n log n)

### ✅ Stable: Yes  
### ❌ In-place: No

```python
def merge_sort(arr):
    if len(arr) <= 1:
        return arr

    mid = len(arr) // 2
    left = merge_sort(arr[:mid])
    right = merge_sort(arr[mid:])

    return merge(left, right)


def merge(left, right):
    result = []
    i = j = 0

    while i < len(left) and j < len(right):
        if left[i] < right[j]:
            result.append(left[i])
            i += 1
        else:
            result.append(right[j])
            j += 1

    result.extend(left[i:])
    result.extend(right[j:])
    return result


arr = [64, 34, 25, 12, 22, 11, 90]
print("Merge Sort:", merge_sort(arr.copy()))
```

## 🔹 Quick Sort

- Picks a pivot and partitions array
- Efficient in practice

### ⏱ Time Complexity:
- Average: O(n log n)
- Worst: O(n²)

### ❌ Stable: No  
### ✅ In-place: Yes (depends on implementation)

```python
def quick_sort(arr):
    if len(arr) <= 1:
        return arr

    pivot = arr[len(arr)//2]
    left = [x for x in arr if x < pivot]
    middle = [x for x in arr if x == pivot]
    right = [x for x in arr if x > pivot]

    return quick_sort(left) + middle + quick_sort(right)


arr = [64, 34, 25, 12, 22, 11, 90]
print("Quick Sort:", quick_sort(arr.copy()))
```
