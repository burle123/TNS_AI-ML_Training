# 📘 Sliding Window Technique

## 🔹 Sliding Window Technique

### 📖 Concept
Sliding Window is an optimization technique used to reduce time complexity.

Instead of recalculating values for every subarray,
we maintain a "window" and slide it across the array.

### ⚡ Key Idea
- Expand window → add element
- Shrink window → remove element

### 🎯 When to Use
- Subarrays / Substrings
- Maximum / Minimum problems
- Continuous segments


## 🔹 Types of Sliding Window

### 1. Fixed Size Window
Window size is constant (e.g., size = k)

### 2. Variable Size Window
Window size changes based on condition


# 🟢 FIXED WINDOW

### Maximum Sum Subarray of Size K

- Maintain window of size k
- Add next element
- Remove first element
- Track maximum sum


```python
def max_sum_subarray(arr, k):
    window_sum = sum(arr[:k])
    max_sum = window_sum

    for i in range(k, len(arr)):
        window_sum += arr[i]
        window_sum -= arr[i - k]
        max_sum = max(max_sum, window_sum)

    return max_sum

print(max_sum_subarray([2,1,5,1,3,2], 5))
```

### First Negative Number in Every Window

Track negative numbers using a queue.


```python
from collections import deque

def first_negative(arr, k):
    result = []
    q = deque()

    for i in range(len(arr)):
        if arr[i] < 0:
            q.append(i)

        if i >= k - 1:
            while q and q[0] < i - k + 1:
                q.popleft()

            result.append(arr[q[0]] if q else 0)

    return result

print(first_negative([12,-1,-7,8,-15,30,16,28], 3))
```

# 🟡 VARIABLE WINDOW

### Longest Substring Without Repeating Characters

- Expand window if no duplicate
- Shrink when duplicate found


```python
def longest_unique_substring(s):
    char_set = set()
    left = 0
    max_length = 0

    for right in range(len(s)):
        while s[right] in char_set:
            char_set.remove(s[left])
            left += 1

        char_set.add(s[right])
        max_length = max(max_length, right - left + 1)

    return max_length

print(longest_unique_substring("abcabcbb"))
```

### Smallest Subarray with Sum ≥ Target

- Expand window to increase sum
- Shrink to minimize length


```python
def min_subarray_len(target, arr):
    left = 0
    window_sum = 0
    min_length = float('inf')

    for right in range(len(arr)):
        window_sum += arr[right]

        while window_sum >= target:
            min_length = min(min_length, right - left + 1)
            window_sum -= arr[left]
            left += 1

    return min_length if min_length != float('inf') else 0

print(min_subarray_len(7, [2,3,1,2,4,3]))
```

### Longest Subarray with Sum ≤ K

Shrink window when sum exceeds K.


```python
def longest_subarray_sum_k(arr, k):
    left = 0
    current_sum = 0
    max_len = 0

    for right in range(len(arr)):
        current_sum += arr[right]

        while current_sum > k:
            current_sum -= arr[left]
            left += 1

        max_len = max(max_len, right - left + 1)

    return max_len

print(longest_subarray_sum_k([1,2,1,0,1,1,0], 4))
```

## 📊 Complexity

- Time Complexity: O(n)
- Space Complexity: O(1) or O(k)

Much better than brute force O(n²)

Sliding Window avoids nested loops.

Convert O(n²) → O(n)
