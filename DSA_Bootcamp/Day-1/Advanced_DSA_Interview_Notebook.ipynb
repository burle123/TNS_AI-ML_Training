# ⏱️ Time & Space Complexity

## 🔸 What is Time Complexity?
Time complexity measures how runtime grows with input size.

### Common Orders:

| Complexity | Meaning | Example |
|----------|--------|--------|
| O(1) | Constant | HashMap lookup |
| O(log n) | Logarithmic | Binary Search |
| O(n) | Linear | Single loop |
| O(n log n) | Efficient sorting | Merge Sort |
| O(n²) | Nested loops | Bubble Sort |

---

## 🔸 Rules to Calculate Complexity

1. Ignore constants → O(2n) = O(n)
2. Keep highest order term → O(n² + n) = O(n²)
3. Nested loops multiply
4. Sequential loops add

---

## 🔸 Space Complexity
Extra memory used by algorithm.

Examples:
- Variables → O(1)
- Array → O(n)
- Recursion → O(stack depth)


```python

import matplotlib.pyplot as plt
import numpy as np

n = np.linspace(1, 100, 100)

plt.plot(n, n, label="O(n)")
plt.plot(n, n*np.log(n), label="O(n log n)")
plt.plot(n, n**2, label="O(n^2)")
plt.plot(n, np.log(n), label="O(log n)")

plt.legend()
plt.title("Time Complexity Comparison")
plt.xlabel("Input Size")
plt.ylabel("Operations")
plt.show()

```


# 👉 Two Pointer Technique

## 🔸 When to Use?
- Sorted arrays
- Pair problems
- Removing duplicates
- Palindrome check

---

## 🔸 Intuition
Instead of nested loops (O(n²)), use two indices.

---

## 🔸 Example: Pair Sum

### Dry Run:
Array = [1,2,3,4,6], Target = 6

left=0, right=4 → sum=7 → move right  
left=0, right=3 → sum=5 → move left  
left=1, right=3 → sum=6 ✅


```python

def two_sum_sorted(arr, target):
    left, right = 0, len(arr)-1

    while left < right:
        current_sum = arr[left] + arr[right]

        if current_sum == target:
            return (left, right)
        elif current_sum < target:
            left += 1
        else:
            right -= 1

    return None

print(two_sum_sorted([1,2,3,4,6], 6))
```

```python

# Remove duplicates (in-place)
def remove_duplicates(arr):
    i = 0
    for j in range(1, len(arr)):
        if arr[i] != arr[j]:
            i += 1
            arr[i] = arr[j]
    return i + 1
```


# 🪟 Sliding Window Technique

## 🔸 When to Use?
- Subarrays / substrings
- Contiguous elements
- Max/min sum problems

---

## 🔸 Types
1. Fixed window (size k)
2. Variable window

---

## 🔸 Example: Max Sum Subarray

### Dry Run:
[2,1,5,1,3,2], k=3

Window moves → update sum efficiently


```python

def max_sum_subarray(arr, k):
    window_sum = sum(arr[:k])
    max_sum = window_sum

    for i in range(k, len(arr)):
        window_sum += arr[i] - arr[i-k]
        max_sum = max(max_sum, window_sum)

    return max_sum

print(max_sum_subarray([2,1,5,1,3,2], 3))
```

```python

# Variable window: longest substring without repeating characters

def longest_unique_substring(s):
    char_set = set()
    left = 0
    max_length = 0

    for right in range(len(s)):
        while s[right] in char_set:
            char_set.remove(s[left])
            left += 1

        char_set.add(s[right])
        max_length = max(max_length, right-left+1)

    return max_length

print(longest_unique_substring("abcabcbb"))
```


# ⚡ Kadane’s Algorithm

## 🔸 Problem
Find maximum subarray sum.

---

## 🔸 Intuition
At each index:
- Either extend previous subarray
- Or start new subarray

---

## 🔸 Dry Run:
[-2,1,-3,4,-1,2,1,-5,4]

Keep updating current and global max


```python

def kadane(arr):
    max_sum = arr[0]
    current_sum = arr[0]

    for i in range(1, len(arr)):
        current_sum = max(arr[i], current_sum + arr[i])
        max_sum = max(max_sum, current_sum)

    return max_sum

print(kadane([-2,1,-3,4,-1,2,1,-5,4]))
```

```python

# Print subarray as well
def kadane_with_subarray(arr):
    max_sum = arr[0]
    current_sum = arr[0]
    start = end = s = 0

    for i in range(1, len(arr)):
        if arr[i] > current_sum + arr[i]:
            current_sum = arr[i]
            s = i
        else:
            current_sum += arr[i]

        if current_sum > max_sum:
            max_sum = current_sum
            start = s
            end = i

    return max_sum, arr[start:end+1]
```
