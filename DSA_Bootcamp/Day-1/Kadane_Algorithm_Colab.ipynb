# 📘 Kadane’s Algorithm

## 🔹 Kadane’s Algorithm

### 📖 Concept
Kadane’s Algorithm is used to find the **maximum sum of a contiguous subarray**.

### ❓ Problem
Given an array, find the subarray with the largest sum.

### ⚡ Key Idea
At each step:
- Either extend the current subarray
- Or start a new subarray

👉 We choose the maximum of both


## ❌ Brute Force Approach

- Generate all subarrays
- Calculate sum for each

### Complexity:
- Time: O(n²)
- Not efficient for large inputs


```python
def max_subarray_bruteforce(arr):
    max_sum = float('-inf')

    for i in range(len(arr)):
        current_sum = 0
        for j in range(i, len(arr)):
            current_sum += arr[j]
            max_sum = max(max_sum, current_sum)

    return max_sum

print(max_subarray_bruteforce([-2,1,-3,4,-1,2,1,-5,4]))
```

## ✅ Kadane’s Algorithm

### ⚙️ Logic

current_sum = max(arr[i], current_sum + arr[i])

max_sum = max(max_sum, current_sum)

### 💡 Intuition
- If adding current element makes sum worse → start fresh
- Otherwise → continue current subarray


```python
def kadane(arr):
    max_sum = arr[0]
    current_sum = arr[0]

    for i in range(1, len(arr)):
        current_sum = max(arr[i], current_sum + arr[i])
        max_sum = max(max_sum, current_sum)

    return max_sum

print(kadane([-2,1,-3,4,-1,2,1,-5,4]))
```

## 🔍 Dry Run

Array:
[-2,1,-3,4,-1,2,1,-5,4]

Final Answer:
[4, -1, 2, 1] → Sum = 6


## 📊 Complexity

- Time Complexity: O(n)
- Space Complexity: O(1)

👉 Huge improvement from O(n²)


### 🔹 Print Maximum Subarray

```python
def kadane_with_subarray(arr):
    max_sum = arr[0]
    current_sum = arr[0]

    start = end = temp_start = 0

    for i in range(1, len(arr)):
        if arr[i] > current_sum + arr[i]:
            current_sum = arr[i]
            temp_start = i
        else:
            current_sum += arr[i]

        if current_sum > max_sum:
            max_sum = current_sum
            start = temp_start
            end = i

    return max_sum, arr[start:end+1]

print(kadane_with_subarray([-2,1,-3,4,-1,2,1,-5,4]))
```

### 🔹 Edge Case: All Negative

```python
print(kadane([-5, -2, -8, -1]))
```
