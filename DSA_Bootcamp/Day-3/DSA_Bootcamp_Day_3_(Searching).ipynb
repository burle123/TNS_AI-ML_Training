## 📖 What is Searching?

Searching is the process of finding a specific element in a dataset (like a list or array).

👉 Example:
Find `10` in `[2, 5, 8, 10, 15]`

---

## 🔹 1. Linear Search

### ✅ Concept
- Traverse the list from start to end
- Compare each element with the target
- Stop when found

### ⏱ Time Complexity
- Worst Case: O(n)

### 👍 When to Use
- Data is **unsorted**
- Small datasets

---

## 🔹 2. Binary Search

### ✅ Concept
- Works on **sorted data**
- Divide the array into halves
- Compare middle element with target

### ⏱ Time Complexity
- O(log n)

### 👍 When to Use
- Large **sorted datasets**

---

## 🔹 3. Jump Search

### ✅ Concept
- Jump ahead by fixed steps
- Once block is found → perform linear search

### ⏱ Time Complexity
- O(√n)

### 👍 When to Use
- Sorted data
- When binary search is not preferred

---

## 📌 Summary

| Algorithm     | Data Requirement | Time Complexity |
|--------------|----------------|----------------|
| Linear Search | Unsorted       | O(n)           |
| Binary Search | Sorted         | O(log n)       |
| Jump Search   | Sorted         | O(√n)          |

#Linear Search

```python
def linear_search(arr, target):
    for i in range(arr):
        if arr[i] == target:
            return i
    return -1

# Example
arr = [4, 2, 7, 1, 9]
target = 7

result = linear_search(arr, target)

if result != -1:
    print("Element found at index:", result)
else:
    print("Element not found")
```

#Binary Search

```python
def binary_search(arr, target):
    low = 0
    high = len(arr) - 1

    while low <= high:
        mid = (low + high) // 2

        if arr[mid] == target:
            return mid
        elif arr[mid] < target:
            low = mid + 1
        else:
            high = mid - 1

    return -1

# Example (Sorted Array)
unsorted_arr = [1, 3, 5, 7, 9, 11,2,4,5,6]
unsorted_arr.sort()
arr = unsorted_arr
print (arr)

target = 5

result = binary_search(arr, target)

if result != -1:
    print("Element found at index:", result)
else:
    print("Element not found")
```

#Jump Search


```python
import math

def jump_search(arr, target):
    n = len(arr)
    step = int(math.sqrt(n))
    prev = 0

    # Finding the block
    while prev < n and arr[min(step, n)-1] < target:
        prev = step
        step += int(math.sqrt(n))
        if prev >= n:
            return -1

    # Linear search within block
    for i in range(prev, min(step, n)):
        if arr[i] == target:
            return i

    return -1

# Example (Sorted Array)
arr = [1, 3, 5, 7, 9, 11, 13, 15]
target = 11

result = jump_search(arr, target)

if result != -1:
    print("Element found at index:", result)
else:
    print("Element not found")
```

#Solution 1 – Linear Search (Find Element Index)

```python
def linear_search(arr, target):
    for i in range(len(arr)):
        if arr[i] == target:
            return i
    return -1

# Example
arr = [10, 20, 30, 40]
target = 30

print(linear_search(arr, target))
```

#Solution 2 – Linear Search (Count Occurrences)

```python
def count_occurrences(arr, target):
    count = 0
    for num in arr:
        if num == target:
            count += 1
    return count

# Example
arr = [1, 2, 3, 2, 2, 4]
target = 2

print(count_occurrences(arr, target))
```

#Solution 3 – Binary Search (Basic Search)

```python
def binary_search(arr, target):
    low = 0
    high = len(arr) - 1

    while low <= high:
        mid = (low + high) // 2

        if arr[mid] == target:
            return mid
        elif arr[mid] < target:
            low = mid + 1
        else:
            high = mid - 1

    return -1

# Example
arr = [1, 3, 5, 7, 9]
target = 5

print(binary_search(arr, target))
```

#Solution 4 – Binary Search (First Occurrence)

```python
def first_occurrence(arr, target):
    low = 0
    high = len(arr) - 1
    result = -1

    while low <= high:
        mid = (low + high) // 2

        if arr[mid] == target:
            result = mid
            high = mid - 1   # move left
        elif arr[mid] < target:
            low = mid + 1
        else:
            high = mid - 1

    return result

# Example
arr = [1, 2, 2, 2, 3, 4]
target = 2

print(first_occurrence(arr, target))
```

```python
def last_occurrence(arr, target):
    low = 0
    high = len(arr) - 1
    result = -1

    while low <= high:
        mid = (low + high) // 2

        if arr[mid] == target:
            result = mid
            low = mid + 1   # move right
        elif arr[mid] < target:
            low = mid + 1
        else:
            high = mid - 1

    return result


# Example
arr = [1, 2, 2, 2, 3, 4]
target = 2

print(last_occurrence(arr, target))
```

#Solution 5 – Jump Search (Basic)

```python
import math

def jump_search(arr, target):
    n = len(arr)
    step = int(math.sqrt(n))
    prev = 0

    while prev < n and arr[min(step, n) - 1] < target:
        prev = step
        step += int(math.sqrt(n))
        if prev >= n:
            return -1

    for i in range(prev, min(step, n)):
        if arr[i] == target:
            return i

    return -1

# Example
arr = [1, 3, 5, 7, 9, 11]
target = 9

print(jump_search(arr, target))
```

Solution 6 – Jump Search (Element Exists or Not)

```python
import math

def jump_search_exists(arr, target):
    n = len(arr)
    step = int(math.sqrt(n))
    prev = 0

    while prev < n and arr[min(step, n) - 1] < target:
        prev = step
        step += int(math.sqrt(n))
        if prev >= n:
            return False

    for i in range(prev, min(step, n)):
        if arr[i] == target:
            return True

    return False

# Example
arr = [2, 4, 6, 8, 10]
target = 6

print(jump_search_exists(arr, target))
```
